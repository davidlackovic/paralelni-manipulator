```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, a, sigma):
    return a * np.exp(-x**2 / (2 * sigma**2))
```

```python
X = np.linspace(-0.2, 0.2, 100)
```

```python
plt.plot(X, gauss(X, 100, 0.02), color='red')
plt.title('Gauss')
plt.axvline(x=-0.18, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.axvline(x=0.18, color='k', linestyle='--')
```
